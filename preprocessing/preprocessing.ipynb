```python
df['explicit'] = df['explicit'].astype(int)
```

```python
feature_cols = [
'danceability',
'energy',
'loudness',
'speechiness',
'acousticness',
'instrumentalness',
'liveness',
'valence',
'tempo',
'popularity',
'duration_ms'
]

features = df[feature_cols]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_features = scaler.fit_transform(features)
```

```python
# Create Fuzzy Mood System

# rule-1
# High energy + high tempo → Energetic

# rule-2
# Low energy + high acousticness → Calm

# rule-3
# High valence → Happy

# rule-4
# Low valence → Sad
```

```python
def fuzzy_mood(row):

    energy = row['energy']
    tempo = row['tempo']
    valence = row['valence']
    acoustic = row['acousticness']

    if energy > 0.7 and tempo > 0.6:
        return "Energetic"

    elif energy < 0.4 and acoustic > 0.5:
        return "Calm"

    elif valence > 0.6:
        return "Happy"

    elif valence < 0.3:
        return "Sad"

    else:
        return "Neutral"
```

```python
df['mood'] = df.apply(fuzzy_mood, axis=1)
```

```python
# Encode Genre
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['genre_encoded'] = le.fit_transform(df['track_genre'])
```

```python
# Create Feature Vectors

import numpy as np

vector_columns = [
'danceability',
'energy',
'loudness',
'speechiness',
'acousticness',
'instrumentalness',
'liveness',
'valence',
'tempo',
'popularity'
]

feature_vectors = df[vector_columns].values
```
